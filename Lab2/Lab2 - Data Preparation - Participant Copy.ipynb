```python
#Import pandas, numpy and maplotlib library 
import pandas as pd
import matplotlib as plt
import numpy as np

#set the filename as "auto.csv", ctreate header list and 
#read the file as df

filename = "auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type",
           "aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", 
           "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system",
           "bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df=pd.read_csv(filename, names=headers)
```

<p>Ways for handle missing values</p>

To replace ? with NaN  in auto.csv, we use .replace(A, B, inplace = True) to replace A by B</p>

<h3>Evaluate Missing Values</h3>
<p>There are two methods available for detecting missing value:

- .isnull()
- .notnull()

The output is a Boolean value indicating whether the value that is passed into the argument is a missing value. </p>

```python
missing_data = df.isnull()
missing_data.head(10)
```

<p>To count the missing values in each column, we use a for loop in Python to quickly figure out the number of missing values in each column. In the body of the for loop the method .value_counts() counts the number of 'True' values.</p>

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
    
```

<b>Replace them with mean</b><br/><br/>
'normalized-losses' - 41 missing values<br/> 
'stroke'           -  4 missing values<br/> 
'bore'             -  4 missing values<br/> 
'horsepower'       -  2 missing values<br/> 
'peak-rpm'         -  2 missing values<br/> 


```python
#code to find mean of normalized-losses and replace the 
#missing values (NaN) in it with mean
avg_norm_loss=df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized losses:", avg_norm_loss)
df["normalized-losses"].replace(np.nan,avg_norm_loss, inplace=True)
#code to find mean of bore and replace the missing values (NaN) 
#in it with mean
avg_bore=df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

<h3>Exercise 2.2</h3>
<p>Write the Python code to find the mean of 'stroke', 'horsepower' and 'peak-rpm' and replace them with means.</p>

```python
avg_stroke=df["stroke"].astype("float").mean(axis=0)
print("Average of stroke:", avg_stroke)
df["stroke"].replace(np.nan,avg_stroke, inplace=True)

avg_horsepower=df["horsepower"].astype("float").mean(axis=0)
print("Average of horsepower:", avg_horsepower)
df["horsepower"].replace(np.nan,avg_horsepower, inplace=True)

avg_peak_rpm=df["peak-rpm"].astype("float").mean(axis=0)
print("Average of peak-rpm:", avg_peak_rpm)
df["peak-rpm"].replace(np.nan,avg_peak_rpm, inplace=True)
```

<h3>Replace them with modes</h3>
<p>- num-of-doors: 2 missing values replace them with 'four'.<br/>

We use .value_counts() to find which values are present in a particular column. We can see that 'four doors' is the most common type by using the .idxmax() method to calculate it. Then replace them with modes</p>

```python
print(df["num-of-doors"].value_counts())
print(df["num-of-doors"].value_counts().idxmax())
#replace the missing 'num-of-doors' values by the mode 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

<h3>Discard them</h3>
<p>price: 	4 missing values (4 rows), simply delete the 4 rows with missing values.</p>


```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

df.head(201)
```

<h3>Data Normalization</h3>

<p>Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable to the range of 0 to 1. <br/>Example min-max normalization = (original value - minimum value)/(maximum value - minimum value)<br/><br/>
Min-max normalization for normalizing the 'length' and 'width' to the ranges of 0 to 1.<br/><br/>
</p>

```python
length_min=df["length"].min()
length_max=df["length"].max()
df["length"] = (df["length"]-length_min)/(length_max - length_min)
width_min=df["width"].min()
width_max=df["width"].max()
df["width"] = (df["width"]-width_min)/(width_max-width_min)
```

<h3>Binning</h3>
<p>It is the process of converting real valued variable into caregorical variables (bins).

Example 'horsepower' ranges from 48 to 288 with 57 unique values. 
If we want to only divide them into 3 categories:</p>
<ul> 
    <li>Low 
    <li>Medium
    <li>High 
</ul>
<p>We use the Pandas cut to segment the 'horsepower' into 3 bins and plot the histogram to see the distribution of 'horsepower'.</p>

```python
df["horsepower"]=df["horsepower"].astype(float, copy=True)
%matplotlib inline
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])
# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

<p>Use numpy's linspace(start_value, end_value, numbers_generated) to generate 3 bins of equal size. </p>

<p>Find the start value and end value of horsepower using min() and max() (start_value=min(df["horsepower"]), end_value=max(df["horsepower"]).Since we are building 3 bins of equal size, there should be 4 dividers, so numbers_generated=4.</p>

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

Set group names and apply 'cut' to determine what bin each value of "df['horsepower']" belongs to.

```python
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, 
                                 labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(201)
```
